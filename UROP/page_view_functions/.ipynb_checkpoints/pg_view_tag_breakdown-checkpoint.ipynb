```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as dates
from matplotlib.ticker import AutoMinorLocator
import datetime as dt
import numpy as np
import gc
```

```python
pg = pd.read_csv('pg_view_1.csv')
```

```python
pg.head()
```

```python
pg['timestamp'] = pg['timestamp'].apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
```

```python
pg.head()
```

```python
df = pd.read_csv('user_movie_view.csv')
del df["userId"]
df2 = pd.read_csv('tags.csv')
del df2['timestampForTag']
del df2['userId']
x = 'action'
df2.drop(df2[df2['tag'] !=x.casefold()].index, inplace=True)
df2 = df2.movieId.unique()
df2 = pd.DataFrame({'movieId':df2})
```

```python
df3 = pd.merge(df2,df,on='movieId')
df3['tstamp'] = df3['tstamp'].apply(lambda x: dt.datetime.strptime(x,'%d-%m-%Y %H:%M'))
```

```python
start = pg['timestamp'].min()
end = pg['timestamp'].max()
df4 = pd.DataFrame(columns=['timestamp','pg_view_'+x])
delta = dt.timedelta(days=1)
i = 0
e = start
print(start)
print(end)
while start <= end:
    count = 0
    e = start+delta
    for t in df3.tstamp:
        if start<=t<e:
            count +=1
    df4.loc[i] = [start,count]
    i+=1
    start = e
```

```python
df_perc = pd.DataFrame(columns=['timestamp','action','christmas','halloween','valentine'])
df_perc['timestamp'] = df_m['timestamp']
df_perc['action'] = df_m['pg_view_action']/(df_m['pg_views']+1)*100
df_perc['christmas'] = df_m['pg_view_christmas']/(df_m['pg_views']+1)*100
df_perc['halloween'] = df_m['pg_view_halloween']/(df_m['pg_views']+1)*100
df_perc['valentine'] = df_m['pg_view_valentine']/(df_m['pg_views']+1)*100
df_perc['comedy'] = df_m['pg_view_comedy']/(df_m['pg_views']+1)*100
df_perc['romamce'] = df_m['pg_view_romance']/(df_m['pg_views']+1)*100

df_perc.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
import matplotlib.dates as mdate
fig.set_size_inches(15,8)

# Major ticks every 20, minor ticks every 5
ax.xaxis_date()
locator = mdate.YearLocator()
plt.gca().xaxis.set_major_locator(locator)

# And a corresponding grid
plt.grid(which='major')

#adding minor ticks
ax.xaxis.set_minor_locator(matplotlib.ticker.AutoMinorLocator(12))
#ax.xaxis.set_minor_formatter(matplotlib.ticker.ScalarFormatter(useMathText=True))
plt.setp(ax.xaxis.get_minorticklabels(), rotation=90)

plt.plot(df_m['timestamp'], df_m['pg_views'],label='views')
plt.plot(df_m['timestamp'], df_m['pg_view_action'],label='action')
plt.plot(df_m['timestamp'], df_m['pg_view_christmas'],label='christmas')
plt.plot(df_m['timestamp'], df_m['pg_view_halloween'],label='halloween')
plt.plot(df_m['timestamp'], df_m['pg_view_valentine'],label='valentine')
plt.plot(df_m['timestamp'], df_m['pg_view_comedy'],label='comedy')
plt.plot(df_m['timestamp'], df_m['pg_view_romance'],label='romance')

plt.legend()
plt.savefig('pg_view_breakdown.pdf')
```
